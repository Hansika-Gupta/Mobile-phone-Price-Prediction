```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
dataset=pd.read_csv("/content/mobiLe data Finale5 - Mobile phone price (2).csv")
dataset
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
print(dataset.columns)
```

```python
dataset.drop(['Model','Storage123','RAM123','Price ($)123','Storage1','Rows','RAM1','rows','Row'],axis=1,inplace=True)
dataset
```

```python

dataset.drop(['Camera (MP)'],axis=1,inplace=True)
dataset

```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
print(X)
print(Y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(sparse_output=False),[0])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
print(X_train)
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
print('Test dataset Y value',Y_test)
Y_pred_lr = regressor.predict(X_test)
print("Predicted",Y_pred_lr)
print("X values",X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

lr_mae = mean_absolute_error(Y_test, Y_pred_lr)
lr_mse = mean_squared_error(Y_test, Y_pred_lr)
lr_r2 = r2_score(Y_test, Y_pred_lr)

print("\nLinear Regression:")
print("MAE:", lr_mae)
print("MSE:", lr_mse)
print("R-squared:", lr_r2)
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
Regressor = KNeighborsRegressor()
Regressor.fit(X_train,Y_train)
```

```python
Y_pred_knn= Regressor.predict(X_test)
print(np.concatenate((Y_pred_knn.reshape(len(Y_pred_knn),1),Y_test.reshape(len(Y_test),1)),1))
```

```python
knn_mae = mean_absolute_error(Y_test, Y_pred_knn)
knn_mse = mean_squared_error(Y_test, Y_pred_knn)
knn_r2 = r2_score(Y_test, Y_pred_knn)

print("\nK-Nearest Neighbors:")
print("MAE:", knn_mae)
print("MSE:", knn_mse)
print("R-squared:", knn_r2)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

#
rf_model = RandomForestRegressor(n_estimators=29, random_state=42)

rf_model.fit(X_train, Y_train)

Y_test_pred = rf_model.predict(X_test)

mse = mean_squared_error(Y_test, Y_test_pred)
mae = mean_absolute_error(Y_test, Y_test_pred)
r2 = r2_score(Y_test, Y_test_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
